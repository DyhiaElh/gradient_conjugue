```python
import matplotlib.pyplot as plt
import pandas as pd

N=80


df_float = pd.read_csv("erreur_float.csv", names=["iteration", "erreur", "residu"])
df_double = pd.read_csv("erreur_double.csv", names=["iteration", "erreur", "residu"])

plt.figure(figsize=(10, 6))

#erreur
plt.plot(df_float["iteration"], df_float["erreur"], label="Erreur (float)", color="blue", linestyle='-')
plt.plot(df_double["iteration"], df_double["erreur"], label="Erreur (double)", color="red", linestyle='-')

#résidu
#plt.plot(df_float["iteration"], df_float["residu"], label="Résidu (float)", color="cyan", linestyle='-.')
#plt.plot(df_double["iteration"], df_double["residu"], label="Résidu (double)", color="orange", linestyle=':')

plt.xlabel("itération")
plt.ylabel("resultats")
plt.title("erreur et résidu")
plt.yscale("log") 

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
